### 4.1.1. Student Engagement Through Attendance

In Section 3 we mentioned student engagement. The most basic metric that can act as a proxy for engagement (excluding minor cases such as medical leave of absence, which are not considered in this analysis) is class attendance. The inverse of this is captured in the dataset under `absences`.

```python
#Create tabulation that compares grades and absences
class_attendance = pd.crosstab(index=df['absences'], columns=[df['grade']], normalize='columns')
#Ignore values where string is empty, as previously set
class_attendance[(class_attendance.index == 'above-7') |(class_attendance.index == 'below-7')].plot(kind='bar',rot=45)
plt.show()
```

From the above, we can see that there is a clear indication that very few high performers have absences larger than 7 days - significantly less than those with medium or low scores. Similarly, as it is expected, a high proportion of high performers have absences below 7 days.

__Insight:__ _Thus, lower levels of absences are present in those students who fall into the high grade category. No major insights can be drawn from the medium and low categories._

### 4.1.2. Grade, Satisfaction and Relationship Pattern Exploration

Parents have a large influence over their offspring. Two columns here are related to this relationship and are intertwined. These are `satisfaction` and `responsible_adult`. Given that these are closely related we will explore them together.

### 4.2.1. Correlation Between Numeric Values and Grade

We will now explore whether there is a correlation between these values and the grade variable.

```python
#Create temp dataframe to alter it without damaging original dataset
df_numeric = []
df_numeric = df[['grade','raised_hand','used_resources','notification_seen','group_participation']]

#Replacing nominal values by numeric ones, to explore correlation
df_numeric.loc[df.grade == 'Low','grade'] = int(0)
df_numeric.loc[df.grade == 'Medium','grade'] = int(50)
df_numeric.loc[df.grade == 'High','grade'] = int(100)

#Convert object to int (as corr only takes numeric values)
df_numeric['grade'] = df_numeric['grade'].astype(int)

#Confirming transformation successful
display(df_numeric['grade'].dtype.kind in 'iufc')

#Plot Pearson correlation, given the Gaussian-like distribution
corr = df_numeric.corr()
sns.heatmap(corr, cmap="binary", annot=True)
```

Although the correlations are not particularly strong given the difficculty of measurement of the variables, as well as the rest of the evidence, there is enough correlation to support the hypothesis that the four metrics (`raised_hand`,`used_resources`,`notification_seen` and `group_participation`) which relate to student engagement are positively correlated with grade. 

`raised_hand` and `used_resources` have very low Pearson correlation coefficients, therefore these will not be further investigated. `notification_seen` and `group_participation` will be explored in more detail.

```python
def new_grade(grade):
    if grade == 'High':
        return 100
    elif grade == 'Medium':
        return 50
    elif grade == 'Low':
        return 10
    
def new_grades(grades): 
    return grades.apply(new_grade)

converted = new_grades(df['grade'])

# Jitter values to see overlaps
x = df['group_participation']
x_jittered = x + 0.6 * np.random.rand(len(x)) 
y = df['notification_seen']
y_jittered = y + 0.6 * np.random.rand(len(y)) 

fig = plt.figure(figsize=(10, 8))
plt.scatter(x_jittered , y_jittered ,c= converted, s = converted, cmap = 'RdYlGn',alpha=0.2)
plt.ylabel("Number of times Student Saw Notifications")
plt.xlabel("Number of times Participated in Group Discussions")
plt.title ('Engagement Correlation by Grades')
plt.colorbar()

##
```

From the above it can be concluded that lower values of engagement (through the proxy of `group_participation` and `notifications_seen`) can be used to predict with relative certitude (r > 0.4). Higher engagement does not necessarily lead to higher results, but lower engagement is good predictor of a student not achieving high results. As can be seen from the figure above. 

Students who were ranked as _low_ did not engage more than 3 times across either of `group_discussions` or ` notification_seen`. Similarly, those ranked as _medium_ did not engage more than 5 times, respectively.

__Insight:__ _Low engagement leads to poor performance, therefore what drives engagement - across these two metrics - should be explored in order to enable students to break free from that poor engagement and potentially lead more successful journeys._
